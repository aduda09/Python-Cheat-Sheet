```python
# Get the lower quartile,q1,median,q2 & upper quartile,q3 =>
import numpy as np
```

```python
# Convert your data into a numpy array
age = np.array([5, 9, 12, 14,15, 17, 19, 21])
```

```python
q1 = np.percentile(age, [25]) # 25th mark=lower quartile,q1;
q1
```

```python
q2 = np.percentile(age, [50]) # 50th mark=Median
q2
```

```python
q3 = np.percentile(age, [75]) # 75th mark=Upper quartile,q1;
q3
```

```python
# IQR = q3 -q1
q3 - q1
```
